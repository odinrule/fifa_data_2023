```python
# grokfucntion
# 定義一個函數來處理字串或整數並轉換為整數
def convert_to_int(value):
    # 如果值是整數，直接返回
    if isinstance(value, (int, float)):
        return int(value)
    
    # 如果值是字串，進行處理
    if isinstance(value, str):
        # 去除歐元符號
        value = value.replace('€', '')
        
        # 檢查是否有 K 或 M，並進行相應的乘法
        if 'K' in value:
            return int(float(value.replace('K', '')) * 1000)
        elif 'M' in value:
            return int(float(value.replace('M', '')) * 1000000)
        else:
            return int(value)
    
    # 如果值是其他類型，返回原值或拋出錯誤
    raise ValueError(f"無法處理的值: {value}")

# 將 Wage 和 Value 欄位應用轉換
df1['Wage'] = df1['Wage'].apply(convert_to_int)
df1['Value'] = df1['Value'].apply(convert_to_int)

# 結果
df1
```

```python
df1['difference'] = df1['Value'] - df1['Wage']
df1.sort_values(by='difference', ascending=False, inplace=True)
df1
```

```python
import seaborn as sns
sns.set()
graph = sns.scatterplot(data=df1, x='Wage', y='Value')
graph
```

```python
from bokeh.plotting import figure, show
from bokeh.models import HoverTool

TOOLTIPS = HoverTool(tooltips=[
    ("index", "$index"),
    ("(Wage, Value)","(@Wage, @Value)"),
    ("Name", "@Name")
])

# 使用 width 和 height 替代 plot_width 和 plot_height
p = figure(title="Soccer 2019", x_axis_label='Wage', y_axis_label='Value', width=700, height=700, tools=[TOOLTIPS])

# 繪製散點圖
p.circle(x='Wage', y='Value', size=10, source=df1)

# 顯示圖表
show(p)
```
